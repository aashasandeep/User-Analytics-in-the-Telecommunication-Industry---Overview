# Aggregate per customer (assuming 'IMSI' is the customer identifier)

```python
aggregated_Telcom_data = Telcom_data.groupby('IMSI').agg({
    'TCP DL Retrans. Vol (Bytes)': 'mean',
    'TCP UL Retrans. Vol (Bytes)': 'mean',
    'Avg RTT DL (ms)': 'mean',
    'Avg RTT UL (ms)': 'mean',
    'Handset Type': lambda x: x.mode()[0],  # most common handset type per customer
    'Avg Bearer TP DL (kbps)': 'mean',
    'Avg Bearer TP UL (kbps)': 'mean'
}).reset_index()
```

# Calculate the average TCP retransmission and RTT

```python
aggregated_Telcom_data['Avg TCP Retrans. Vol (Bytes)'] = (aggregated_Telcom_data['TCP DL Retrans. Vol (Bytes)'] +
                                                 aggregated_Telcom_data['TCP UL Retrans. Vol (Bytes)']) / 2
```

```python
aggregated_Telcom_data['Avg TCP Retrans. Vol (Bytes)']
```

```python
aggregated_Telcom_data['Avg RTT (ms)'] = (aggregated_Telcom_data['Avg RTT DL (ms)'] +
                                 aggregated_Telcom_data['Avg RTT UL (ms)']) / 2
```

```python
aggregated_Telcom_data['Avg RTT (ms)']
```

```python
aggregated_Telcom_data['Avg Throughput (kbps)'] = (aggregated_Telcom_data['Avg Bearer TP DL (kbps)'] +
                                          aggregated_Telcom_data['Avg Bearer TP UL (kbps)']) / 2
```

```python
aggregated_Telcom_data['Avg Throughput (kbps)']
```

# Drop unnecessary columns

```python
aggregated_Telcom_data.drop(columns=['TCP DL Retrans. Vol (Bytes)', 'TCP UL Retrans. Vol (Bytes)',
                            'Avg RTT DL (ms)', 'Avg RTT UL (ms)', 'Avg Bearer TP DL (kbps)',
                            'Avg Bearer TP UL (kbps)'], inplace=True)
```

# Define a function to get top, bottom, and most frequent values

```python
def get_top_bottom_most_frequent(aggregated_Telcom_data, col, top_n=10):
    top_values = aggregated_Telcom_data[col].nlargest(top_n)
    bottom_values = aggregated_Telcom_data[col].nsmallest(top_n)
    most_frequent_values = aggregated_Telcom_data[col].value_counts().nlargest(top_n)
    return top_values, bottom_values, most_frequent_values
```

```python
top_tcp, bottom_tcp, most_freq_tcp = get_top_bottom_most_frequent(aggregated_Telcom_data, 'Avg TCP Retrans. Vol (Bytes)')
```

```python
top_tcp, bottom_tcp, most_freq_tcp
```

```python
top_rtt, bottom_rtt, most_freq_rtt = get_top_bottom_most_frequent(aggregated_Telcom_data, 'Avg RTT (ms)')
```

```python
top_rtt, bottom_rtt, most_freq_rtt
```

```python
top_throughput, bottom_throughput, most_freq_throughput = get_top_bottom_most_frequent(aggregated_Telcom_data, 'Avg Throughput (kbps)')
```

```python
top_throughput, bottom_throughput, most_freq_throughput
```

```python
plt.figure(figsize=(15, 8))
sns.barplot(data=throughput_per_handset, x='Most Common Handset Type', y='Avg Throughput (kbps)')
plt.xticks(rotation=90)
plt.title('Average Throughput per Most Common Handset Type')
plt.xlabel('Handset Type')
plt.ylabel('Average Throughput (kbps)')
plt.tight_layout()
plt.show()
```

# Handset types with the highest average throughout are typically newer and higher-end models with better hardware and software optimizations for network performance.

Handset types with the lowest average throughout may be older models or budget devices with limited network capabilities.

# Select relevant features for clustering

```python
features = aggregated_Telcom_data[['Avg TCP Retrans. Vol (Bytes)', 'Avg RTT (ms)', 'Avg Throughput (kbps)']]
```

```python
features
```

# Standardize the data

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)
```

# Perform k-means clustering

```python
kmeans = KMeans(n_clusters=3, random_state=42)
aggregated_Telcom_data['Cluster'] = kmeans.fit_predict(features_scaled)
```

# Analyze and describe the clusters

```python
cluster_centers = kmeans.cluster_centers_
cluster_centers = scaler.inverse_transform(cluster_centers)
```

# Add cluster centers to the DataFrame for interpretation

```python
cluster_summary = pd.DataFrame(cluster_centers, columns=['Avg TCP Retrans. Vol (Bytes)', 'Avg RTT (ms)', 'Avg Throughput (kbps)'])
cluster_summary['Cluster'] = range(3)
```

```python
cluster_summary
```

# Plot the distribution of each cluster

```python
for i, col in enumerate(['Avg TCP Retrans. Vol (Bytes)', 'Avg RTT (ms)', 'Avg Throughput (kbps)']):
    plt.subplot(1, 3, i+1)
    sns.boxplot(x='Cluster', y=col, data=aggregated_Telcom_data)
    plt.title(f'{col} by Cluster')

plt.tight_layout()
plt.show()
```
